```python
import pandas as pd
```

```python
import os
print(os.path.exists('../data/raw_leads.csv'))
```

```python
df = pd.read_csv('../data/raw_leads.csv')
```

```python
# Convert to datetime objects
df['Lead_Date'] = pd.to_datetime(df['Lead_Date'])
df['Conversion_Date'] = pd.to_datetime(df['Conversion_Date'])
```

```python
# Calculate Sales Cycle Duration early (Day 1 Task)
df['Sales_Cycle_Days'] = (df['Conversion_Date'] - df['Lead_Date']).dt.days
```

```python
# Export Cleaned Data
df.to_csv('../data/cleaned_leads.csv', index=False)
```

```python
# --- DATA VALIDATION CHECK ---

# 1. Check for missing values in critical columns
# These columns MUST have 0 missing values
critical_cols = ['Lead_ID', 'Lead_Source', 'Lead_Date', 'Conversion_Status', 
                 'Marketing_Cost', 'Sales_Representative', 'Region']

missing_critical = df[critical_cols].isnull().sum()
print("1. Missing values in mandatory columns (Should be 0):")
print(missing_critical)

# 2. Logic Check: Revenue for 'Not Converted' leads
# These should all be 0.0
non_converted_rev = df[df['Conversion_Status'] == 'Not Converted']['Revenue_Value'].sum()
print(f"\n2. Total Revenue from 'Not Converted' leads: {non_converted_rev} (Should be 0)")

# 3. Logic Check: Conversion Dates
# Number of Converted leads should match number of Conversion_Dates
converted_count = len(df[df['Conversion_Status'] == 'Converted'])
conv_date_count = df['Conversion_Date'].notnull().sum()
print(f"\n3. Converted Leads: {converted_count} | Conversion Dates found: {conv_date_count}")

# 4. Final Data Types Check
print("\n4. Column Data Types (Ensure dates are datetime64):")
print(df.dtypes)

if (missing_critical.sum() == 0) and (converted_count == conv_date_count):
    print("\n✅ DATA IS CLEAN: Ready for Day 2 KPI Analysis.")
else:
    print("\n❌ DATA ISSUE FOUND: Please re-run cleaning steps.")
```
